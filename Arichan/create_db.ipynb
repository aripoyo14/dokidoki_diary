```python
import sqlite3

# SQLiteデータベースを作成
conn = sqlite3.connect('dokidoki_diary.db')
c = conn.cursor()

# テーブルを作成
c.execute('''
          CREATE TABLE Diary_table(
              diary_id INTEGER PRIMARY KEY AUTOINCREMENT,
              user_id INT,
              date date,
              diary TEXT,
              feedback TEXT
          )
          ''')

conn.commit()
conn.close()
```
